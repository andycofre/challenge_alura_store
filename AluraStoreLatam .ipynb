### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_total_tienda1 = tienda ["Precio"].sum()
facturacion_total_tienda2 = tienda2 ["Precio"].sum()
facturacion_total_tienda3 = tienda3 ["Precio"].sum()
facturacion_total_tienda4 = tienda4 ["Precio"].sum()

print(f"Facturación total tienda 1: {facturacion_total_tienda1}")
print(f"Facturación total tienda 2: {facturacion_total_tienda2}")
print(f"Facturación total tienda 3: {facturacion_total_tienda3}")
print(f"Facturación total tienda 4: {facturacion_total_tienda4}")
```

```python
import matplotlib.pyplot as plt

# Datos
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
facturaciones = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

# Lista de colores (puedes cambiar estos a los que más te gusten)
colores = ['#FF5733', '#33C1FF', '#75FF33', '#FFC300']  # Rojo, azul, verde, amarillo

# Gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, facturaciones, color=colores)
plt.title("Facturación Total por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Facturación")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
print(tienda.columns)
print(tienda2.columns)
print(tienda3.columns)
print(tienda4.columns)
```

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    print(f"\n📊 {nombre} - Categorías más populares (por número de ventas):")
    conteo = df.groupby("Categoría del Producto").size().sort_values(ascending=False)
    print(conteo)
```

```python
import matplotlib.pyplot as plt

for nombre, df in tiendas.items():
    conteo = df["Categoría del Producto"].value_counts()

    plt.figure(figsize=(7, 7))
    plt.pie(conteo, labels=conteo.index, autopct='%1.1f%%', startangle=140)
    plt.title(f"{nombre} - Categorías más vendidas (en %)")
    plt.axis('equal')  # Mantiene el círculo perfecto
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

print(tienda.head())
print(tienda.columns)

```

```python
# Conversión a numérico (por si acaso en otros hay texto)
for df in [tienda, tienda2, tienda3, tienda4]:
    df["Calificación"] = pd.to_numeric(df["Calificación"], errors="coerce")

# Calcular promedios
calif_prom_t1 = tienda["Calificación"].mean()
calif_prom_t2 = tienda2["Calificación"].mean()
calif_prom_t3 = tienda3["Calificación"].mean()
calif_prom_t4 = tienda4["Calificación"].mean()

# Mostrar resultados
print(f" Calificación promedio Tienda 1: {calif_prom_t1:.2f}")
print(f" Calificación promedio Tienda 2: {calif_prom_t2:.2f}")
print(f" Calificación promedio Tienda 3: {calif_prom_t3:.2f}")
print(f" Calificación promedio Tienda 4: {calif_prom_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Nombres de las tiendas y sus calificaciones promedio
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
promedios = [calif_prom_t1, calif_prom_t2, calif_prom_t3, calif_prom_t4]
colores = ["#66c2a5", "#fc8d62", "#8da0cb", "#e78ac3"]  # Colores bonitos y diferenciados

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
bars = plt.bar(nombres_tiendas, promedios, color=colores)

# Añadir etiquetas de valor encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.1, f"{yval:.2f}", ha='center', va='bottom')

# Títulos y etiquetas
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación (1 a 5)")
plt.ylim(0, 5)  # Escala estándar de calificación
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

print(tienda.head())
print(tienda.columns)
```

```python


# Tienda 1 - Top 5 productos más vendidos
top5_t1 = tienda["Producto"].value_counts().head(5)

# Tienda 2 - Top 5 productos más vendidos
top5_t2 = tienda2["Producto"].value_counts().head(5)

# Tienda 3 - Top 5 productos más vendidos
top5_t3 = tienda3["Producto"].value_counts().head(5)

# Tienda 4 - Top 5 productos más vendidos
top5_t4 = tienda4["Producto"].value_counts().head(5)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Tienda 1
axs[0, 0].bar(top5_t1.index, top5_t1.values, color='#66c2a5')
axs[0, 0].set_title("Tienda 1")

# Tienda 2
axs[0, 1].bar(top5_t2.index, top5_t2.values, color='#fc8d62')
axs[0, 1].set_title("Tienda 2")

# Tienda 3
axs[1, 0].bar(top5_t3.index, top5_t3.values, color='#8da0cb')
axs[1, 0].set_title("Tienda 3")

# Tienda 4
axs[1, 1].bar(top5_t4.index, top5_t4.values, color='#e78ac3')
axs[1, 1].set_title("Tienda 4")

# Ajustes generales
for ax in axs.flat:
    ax.set_ylabel("Ventas")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.suptitle("Top 5 productos más vendidos por tienda", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Tienda 1 - 5 productos menos vendidos
menos_vendidos_t1 = tienda["Producto"].value_counts().tail(5)

# Tienda 2 - 5 productos menos vendidos
menos_vendidos_t2 = tienda2["Producto"].value_counts().tail(5)

# Tienda 3 - 5 productos menos vendidos
menos_vendidos_t3 = tienda3["Producto"].value_counts().tail(5)

# Tienda 4 - 5 productos menos vendidos
menos_vendidos_t4 = tienda4["Producto"].value_counts().tail(5)

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Tienda 1
axs[0, 0].bar(menos_vendidos_t1.index, menos_vendidos_t1.values, color='#ff9999')
axs[0, 0].set_title("Tienda 1")

# Tienda 2
axs[0, 1].bar(menos_vendidos_t2.index, menos_vendidos_t2.values, color='#ffcc99')
axs[0, 1].set_title("Tienda 2")

# Tienda 3
axs[1, 0].bar(menos_vendidos_t3.index, menos_vendidos_t3.values, color='#c2c2f0')
axs[1, 0].set_title("Tienda 3")

# Tienda 4
axs[1, 1].bar(menos_vendidos_t4.index, menos_vendidos_t4.values, color='#ffb3e6')
axs[1, 1].set_title("Tienda 4")

# Ajustes generales
for ax in axs.flat:
    ax.set_ylabel("Ventas")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.suptitle("Top 5 productos menos vendidos por tienda", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
costo_envio_prom_t1 = tienda["Costo de envío"].mean()
costo_envio_prom_t2 = tienda2["Costo de envío"].mean()
costo_envio_prom_t3 = tienda3["Costo de envío"].mean()
costo_envio_prom_t4 = tienda4["Costo de envío"].mean()

print(f"Costo de envío promedio Tienda 1: {costo_envio_prom_t1:.2f}")
print(f"Costo de envío promedio Tienda 2: {costo_envio_prom_t2:.2f}")
print(f"Costo de envío promedio Tienda 3: {costo_envio_prom_t3:.2f}")
print(f"Costo de envío promedio Tienda 4: {costo_envio_prom_t4:.2f}")

```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
costos_envio = [costo_envio_prom_t1, costo_envio_prom_t2, costo_envio_prom_t3, costo_envio_prom_t4]

# Crear gráfico de líneas
plt.figure(figsize=(8, 5))
plt.plot(tiendas, costos_envio, marker='o', linestyle='-', color='teal', linewidth=2)

# Etiquetas de los valores sobre los puntos
for i, valor in enumerate(costos_envio):
    plt.text(i, valor + 500, f"${valor:,.2f}", ha='center', va='bottom', fontsize=9)

# Títulos y etiquetas
plt.title("Costo de envío promedio por tienda", fontsize=14)
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
